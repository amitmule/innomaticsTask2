```python
class Solution:
    def subtractProductAndSum(self, n: int) -> int:
        prod = 1
        summ = 0
        for digit in str(n):
            d = int(digit)
            prod *= d
            summ += d
        return prod - summ

# Instantiate
sol = Solution()

# Example 1
n1 = 234
# Product: 2*3*4 = 24, Sum: 2+3+4 = 9, Result: 15
print(f"Input: {n1} -> Output: {sol.subtractProductAndSum(n1)}")

# Example 2
n2 = 4421
# Product: 4*4*2*1 = 32, Sum: 4+4+2+1 = 11, Result: 21
print(f"Input: {n2} -> Output: {sol.subtractProductAndSum(n2)}")
```
