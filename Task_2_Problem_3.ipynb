```python
from typing import List

class Solution:
    def numIdenticalPairs(self, nums: List[int]) -> int:
        seen = {}
        pairs = 0
        for num in nums:
            if num in seen:
                pairs += seen[num]
                seen[num] += 1
            else:
                seen[num] = 1
        return pairs

sol = Solution()

# Example 1
nums1 = [1, 2, 3, 1, 1, 3]
print(f"Ex 1 Input: {nums1} -> Output: {sol.numIdenticalPairs(nums1)}")
# Expected: 4

# Example 2
nums2 = [1, 1, 1, 1]
print(f"Ex 2 Input: {nums2} -> Output: {sol.numIdenticalPairs(nums2)}")
# Expected: 6

# Example 3 (The one I missed!)
nums3 = [1, 2, 3]
print(f"Ex 3 Input: {nums3} -> Output: {sol.numIdenticalPairs(nums3)}")
# Expected: 0
```
