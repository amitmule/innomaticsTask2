```python
from typing import List

class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        # Sort to find the 'rank' of each number
        sorted_nums = sorted(nums)
        rank = {}

        for i, num in enumerate(sorted_nums):
            if num not in rank:
                rank[num] = i

        return [rank[num] for num in nums]

# Instantiate
sol = Solution()

# Example 1
nums1 = [8, 1, 2, 2, 3]
print(f"Ex 1 Input: {nums1}")
print(f"Output: {sol.smallerNumbersThanCurrent(nums1)}")
# Expected: [4, 0, 1, 1, 3]

# Example 2
nums2 = [6, 5, 4, 8]
print(f"\nEx 2 Input: {nums2}")
print(f"Output: {sol.smallerNumbersThanCurrent(nums2)}")
# Expected: [2, 1, 0, 3]

# Example 3
nums3 = [7, 7, 7, 7]
print(f"\nEx 3 Input: {nums3}")
print(f"Output: {sol.smallerNumbersThanCurrent(nums3)}")
# Expected: [0, 0, 0, 0]
```
