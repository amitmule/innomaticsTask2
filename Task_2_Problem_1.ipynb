```python
# 1. Define the Class
class Solution:
    def defangIPaddr(self, address: str) -> str:
        return address.replace('.', '[.]')

# 2. Create an instance of the class
sol = Solution()

# 3. Test with Example 1
input_1 = "1.1.1.1"
output_1 = sol.defangIPaddr(input_1)
print(f"Input: {input_1}")
print(f"Output: {output_1}")

# 4. Test with Example 2
input_2 = "255.100.50.0"
output_2 = sol.defangIPaddr(input_2)
print(f"Input: {input_2}")
print(f"Output: {output_2}")
```
