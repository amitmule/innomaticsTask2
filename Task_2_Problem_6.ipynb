```python
from typing import List

class Solution:
    def xorOperation(self, n: int, start: int) -> int:
        result = 0
        for i in range(n):
            # Calculate the term based on the formula
            term = start + 2 * i
            # XOR it with the current total
            result ^= term
        return result

# Instantiate
sol = Solution()

# Example 1
n1, start1 = 5, 0
# Array: [0, 2, 4, 6, 8] -> 0^2^4^6^8 = 8
print(f"n={n1}, start={start1} -> Output: {sol.xorOperation(n1, start1)}")

# Example 2
n2, start2 = 4, 3
# Array: [3, 5, 7, 9] -> 3^5^7^9 = 8
print(f"n={n2}, start={start2} -> Output: {sol.xorOperation(n2, start2)}")
```
