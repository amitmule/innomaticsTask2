```python
from typing import List
import math

# 1. Define the class
class Solution:
    def findNumbers(self, nums: List[int]) -> int:
        count = 0
        for num in nums:
            if len(str(num)) % 2 == 0:
                count += 1
        return count

# 2. Instantiate and Test
sol = Solution()

# Test Case 1
nums1 = [12, 345, 2, 6, 7896]
print(f"Input: {nums1}")
print(f"Output: {sol.findNumbers(nums1)}") # Expected: 2

# Test Case 2
nums2 = [555, 901, 482, 1771]
print(f"Input: {nums2}")
print(f"Output: {sol.findNumbers(nums2)}") # Expected: 1
```
